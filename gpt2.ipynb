```python
import transformers
from transformers import pipeline
```

```python
text_gen = pipeline("text-generation", model="openai-community/gpt2")
```

```python
request = "GPT (Generative Pre-trained Transformer) is a deep learning model developed by OpenAI that excels at understanding and generating human-like text"
result = text_gen(request, max_length=600, truncation=False, do_sample=True)
print(result)
```
